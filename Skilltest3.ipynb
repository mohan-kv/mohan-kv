```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

X=data.drop(columns='Salary')
y=data['Salary']

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)

model1=LinearRegression()
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print('Linear Regression: \n',pred1)
```

```python
model1=DecisionTreeRegressor()
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print('Decision Tree Regression: \n',pred1)
```

```python
from sklearn.svm import SVC
model11=SVC()
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print('Binary Classification: \n',pred1)
```

```python
from sklearn.ensemble import RandomForestClassifier

model1=RandomForestClassifier()
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print('Multi Label Classifiaction : \n',pred1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
pred1=model1.predict(x_test)
print('Multi Class Classification: \n',pred1)
```

```python
from sklearn.naive_bayes import GaussianNB

mode12=GaussianNB()
mode12.fit(x_train,y_train)
predd=mode12.predict(x_test)
print('Imbalanced Classifiaction : \n',predd)
```

Accuracy,Precison,etc

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

ac=accuracy_score(y_test,model1.predict(x_test))
print('accuracy',ac)
```

```python
pr=precision_score(y_test,mode12.predict(x_test),average='weighted')
print('precision',pr)
```

```python
recall=recall_score(y_test,mode12.predict(x_test),average='weighted')
```

```python
f1=f1_score(y_test,model1.predict(x_test),average='weighted')
print(f1)
```

```python
cm=confusion_matrix(y_test,model1.predict(x_test))
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(cm,annot=True,cmap='coolwarm')
plt.title('Confusion Matrix')
plt.show()
```

```python
spec=cm/cm+cm
print('specificity',spec)
```
