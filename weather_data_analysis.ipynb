```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    'Time':['2024-07-27 00:00', '2024-07-27 01:00', '2024-07-27 02:00','2024-07-27 03:00', '2024-07-27 04:00', '2024-07-27 05:00',
            '2024-07-27 06:00', '2024-07-27 07:00', '2024-07-27 08:00','2024-07-27 09:00', '2024-07-27 10:00', '2024-07-27 11:00',
            '2024-07-27 12:00', '2024-07-27 13:00', '2024-07-27 14:00','2024-07-27 15:00', '2024-07-27 16:00', '2024-07-27 17:00',
            '2024-07-27 18:00', '2024-07-27 19:00', '2024-07-27 20:00','2024-07-27 21:00', '2024-07-27 22:00', '2024-07-27 23:00'],
    'Temperature':[20, 19, 18, 17, 16, 15, 15, 16, 18, 20, 22, 24,25, 26, 27, 28, 29, 30, 29, 27, 25, 24, 22, 21],
    'Humidity':[90, 92, 93, 94, 95, 96, 97, 96, 94, 92, 90, 88,87, 85, 82, 80, 78, 75, 77, 80, 82, 84, 86, 88],
    'Wind Speed':[5, 4, 3, 3, 2, 1, 1, 2, 3, 4, 5, 5,6, 7, 6, 5, 4, 3, 3, 2, 2, 3, 4, 5]
}

df=pd.DataFrame(data)
df['Time']=pd.to_datetime(df['Time'])

plt.figure(figsize=(15,10))
plt.subplot(3,1,1)
plt.plot(df['Time'],df['Temperature'],color='r',marker='o',linestyle='-',label='Temperature (C)')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.subplot(3,1,2)
plt.plot(df['Time'],df['Humidity'],color='b',marker='o',linestyle='-',label='Humidity (%)')
plt.title('Humidity Over Time')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

plt.subplot(3,1,3)
plt.plot(df['Time'],df['Wind Speed'],color='g',marker='o',linestyle='-',label='Wind Speed (km/h)')
plt.title('Wind Speed Over Time')
plt.xlabel('Time')
plt.ylabel('Wind Speed (km/h)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
