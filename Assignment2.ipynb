**calculated and prints the mean,mode,median,standard deviation,variance, range,sum,count,minimum, and maximum of all columns in a dataset**

```python
data['Mat']=pd.to_numeric(data['Mat'],errors='coerce')
data['Inns']=pd.to_numeric(data['Inns'],errors='coerce')
data['NO']=pd.to_numeric(data['NO'],errors='coerce')
data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
data['HS']=pd.to_numeric(data['HS'],errors='coerce')
data['Ave']=pd.to_numeric(data['Ave'],errors='coerce')
data['BF']=pd.to_numeric(data['BF'],errors='coerce')
data['SR']=pd.to_numeric(data['SR'],errors='coerce')
data['100']=pd.to_numeric(data['100'],errors='coerce')
data['50']=pd.to_numeric(data['50'],errors='coerce')
data['0']=pd.to_numeric(data['0'],errors='coerce')
data['4s']=pd.to_numeric(data['4s'],errors='coerce')
data['6s']=pd.to_numeric(data['6s'],errors='coerce')

numeric=['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']
mean1=data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']].mean()
mode1=data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']].mode().iloc[0]
median1=data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']].median()
std1=data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']].std()
var1=data[['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0', '4s', '6s']].var()
range1=data[numeric].max() - data[numeric].min()
sum1=data[numeric].sum()
count1=data[numeric].count()
min1=data[numeric].min()
max1=data[numeric].max()

print("Mean values:\n",mean1)
print("\nMode values:\n",mode1)
print("\nMedian values:\n",median1)
print("\nStandard Deviation values:\n",std1)
print("\nVariance values:\n",var1)
print("\nRange Values:\n",range1)
print("\nSum Values:\n",sum1)
print("\nCount Values:\n",count1)
print("\nMinimum values:\n",min1)
print("\nMaximum Values:\n",max1)
```

**Univariate tests for the dataset.**

**One Sample Test**

```python
import scipy.stats as stats

data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
cr=data['Runs'].dropna()

t_statistic,p_value=stats.ttest_1samp(cr,5.9)
print("t-statistic:",t_statistic)
print("p-value:",p_value)
```

**Chi-Square Test**

```python
subset=data[['Mat', 'Inns', 'Runs']]
subset_numeric=subset.apply(pd.to_numeric, errors='coerce').dropna()

chi2,p,dof,expected=stats.chi2_contingency(subset_numeric)
print(f"Chi-Square Test:chi2={chi2},p-value={p},dof={dof}")
```

**one-way ANOVA (Analysis of Variance)**

```python
data['Inns']=pd.to_numeric(data['Inns'],errors='coerce')
data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
datac= data.dropna(subset=['Inns','Runs'])

f_statistic, p_value = stats.f_oneway(
    datac['Runs'][datac['Inns'] < 30],
    datac['Runs'][(datac['Inns'] >= 30) & (datac['Inns'] < 50)],
    datac['Runs'][datac['Inns'] >= 50])
print(f"F-statistic: {f_statistic}, p-value: {p_value}")
```

**Mann-Whitney**

```python
data['Inns']=pd.to_numeric(data['Inns'],errors='coerce')
data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
datac= data.dropna(subset=['Inns','Runs'])

younger=datac[datac['Inns'] < 30]['Runs']
older=datac[datac['Inns'] >= 30]['Runs']

stat,p_val=stats.mannwhitneyu(younger,older)
print(f"Mann-Whitney U Test:U={stat},p={p_val}")
```

**Kruskal-Wallis Test**

```python
data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
data=data.dropna(subset=['Runs'])

t1=data[data['Player'].str.contains('INDIA',na=False)]['Runs']
t2=data[data['Player'].str.contains('NZ',na=False)]['Runs']

stat,p_val=stats.kruskal(t1,t2)
print(f"Kruskal-Wallis Test: H-statistic={stat},p-value={p_val}")
```

**Univariate distribution plots**

Univariate distribution plots are used to visualize the distribution of a single variable.        
                                                  Common types include histograms, kernel density plots, and box plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
data = data.dropna(subset=['Runs'])

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.hist(data['Runs'],bins=10,edgecolor='black',alpha=0.7)
plt.title('Histogram of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')

plt.subplot(1,3,2)
sns.kdeplot(data['Runs'],shade=True,color='blue')
plt.title('Kernel Density Plot of Runs')
plt.xlabel('Runs')
plt.ylabel('Density')

plt.subplot(1,3,3)
sns.boxplot(data['Runs'])
plt.title('Box Plot of Runs')
plt.xlabel('Runs')
plt.tight_layout()
plt.show()

```

**Univariate comparison plots**

Univariate comparison plots allow you to compare the distribution of a single variable across different categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data['Runs']=pd.to_numeric(data['Runs'],errors='coerce')
data=data.dropna(subset=['Runs'])

plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.histplot(data,x='Runs',hue='Mat',palette='Set2',multiple='stack',alpha=0.7)
plt.title('Histogram of Runs by Match')
plt.xlabel('Runs')
plt.ylabel('Frequency')

plt.subplot(2,2,2)
sns.kdeplot(data, x='Runs',hue='Mat',palette='Set2',multiple='stack',shade=True)
plt.title('Kernel Density Plot of Runs by Match')
plt.xlabel('Runs')
plt.ylabel('Density')

plt.subplot(2,2,3)
sns.boxplot(data=data,x='Mat',y='Runs',palette='Set2')
plt.title('Box Plot of Runs by Match')
plt.xlabel('Match')
plt.ylabel('Runs')
plt.xticks(rotation=90)

plt.subplot(2,2,4)
sns.violinplot(data=data,x='Mat',y='Runs',inner='quartile',palette='Set2')
plt.title('Violin Plot of Runs by Match')
plt.xlabel('Match')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# **5.Perform any probability calculation**

```python
import pandas as pd

data['100']=pd.to_numeric(data['100'],errors='coerce')
data['50']=pd.to_numeric(data['50'],errors='coerce')
data['4s']=pd.to_numeric(data['4s'],errors='coerce')
data['6s']=pd.to_numeric(data['6s'],errors='coerce')

total_players=len(data)
pcentury=len(data[data['100']>0])/total_players
pfifty=len(data[data['50']>0])/total_players
pfours=len(data[data['4s']>100])/total_players
psixes=len(data[data['6s']>50])/total_players

print(f"Probability of scoring a century:{pcentury:.2f}")
print(f"Probability of scoring a half-century:{pfifty:.2f}")
print(f"Probability of hitting more than 100 fours:{pfours:.2f}")
print(f"Probability of hitting more than 50 sixes:{psixes:.2f}")
```
