```python
import pandas as pd

data=pd.read_csv('/content/results.csv')
print(data)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report

x=data[['Hindi','English','Science','Maths','History','Geograpgy']]
y=data['Total']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model1=LogisticRegression(max_iter=1000)
model1.fit(x_train,y_train)
res=model1.predict(x_test)
print('Logistic Regression: \n',res)
print('classification report: ',classification_report(y_test,res))

model2=SVC()
model2.fit(x_train,y_train)
res1=model2.predict(x_test)
print('support vector machine: \n',res1)
print('classifiaction report: ',classification_report(y_test,res1))
```

```python
from sklearn.cluster import KMeans

sin=pd.get_dummies(data)
sin1=KMeans(n_clusters=3).fit(sin)
result=sin1.inertia_
print("K-means Result",result)
```

```python
from sklearn.cluster import cluster_centers_

print(result.cluster_centers_)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

k=range(1,11)
elbow=[KMeans(n_clusters=k,random_state=0).fit(x).inertia_ for k in k]
print(elbow)
plt.plot(k,elbow)
plt.show()
```
