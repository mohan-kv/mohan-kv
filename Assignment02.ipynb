```python
import pandas as pd
data=pd.read_csv('/content/1000_Companies.csv')
print(data)
```

**1 Analyze the univariate dataset Ex- Mean, Mode, Median,**
**Range, Std, and Variance and perform Univariate tests**
**for the dataset.**

```python
d1=data.apply(pd.to_numeric, errors='coerce')
mean=d1.mean()
mode=d1.mode().iloc[0]
median=d1.median()
count=d1.count()
drange=d1.max()-data.min()
std=d1.std()
variance=d1.var()
minimum=d1.min()
maximum=d1.max()

print("Mean:\n",mean)
print("\nMode:\n",mode)
print("\nMedian:\n",median)
print("\nCount:\n",count)
print("\nRange:\n",drange)
print("\nStandard Deviation:\n",std)
print("\nVariance:\n",variance)
print("\nMinimum\n",minimum)
print("\nMaximum\n",maximum)
```

```python
import scipy.stats as stats

data['R&D Spend']=pd.to_numeric(data['R&D Spend'],errors='coerce')
cleaned_runs=data['R&D Spend'].dropna()

t_statistic,p_value=stats.ttest_1samp(cleaned_runs,5.9)
print("t-statistic:",t_statistic)
print("p-value:",p_value)
```

```python
from scipy.stats import normaltest

stat, p_value = normaltest(data)
print(f"Normality Test Statistic:{stat},p-value:{p_value}")
```

```python
sub=data[['R&D Spend', 'Profit','Marketing Spend']]
subnum=sub.apply(pd.to_numeric,errors='coerce').dropna()

chi2,p,dof,expected=stats.chi2_contingency(subnum)
print(f"Chi-Square Test:chi2={chi2},p-value={p},dof={dof}")
```

```python
df = pd.read_csv('1000_Companies.csv').dropna(subset=['Profit', 'State'])
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

groups = [df[df['State'] == state]['Profit'] for state in df['State'].unique()]
print("one Way Test: ",stats.f_oneway(*groups))
```

```python
print("Kruskal-Wallis Test: ",stats.kruskal(*groups))
```

# **Univariate Distribution Plots**

Histogram:Shows the distribution of profit
Box Plot:Displays the spread and outliers in profit
KDE Plot:Kernel Density Estimate plot of profit

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data[['R&D Spend','Administration','Marketing Spend','Profit']] = df[['R&D Spend','Administration','Marketing Spend','Profit']].apply(pd.to_numeric, errors='coerce')
dc= df.dropna(subset=['R&D Spend', 'Administration', 'Marketing Spend', 'Profit'])

sns.set(style='whitegrid')
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.histplot(dc['Profit'],kde=True)
plt.title('Histogram of Profit')
plt.subplot(1, 3, 2)
sns.boxplot(dc['Profit'])
plt.title('Box Plot of Profit')

plt.subplot(1, 3, 3)
sns.kdeplot(dc['Profit'])
plt.title('KDE Plot of Profit')
plt.tight_layout()
plt.show()
```

# **Univariate Comparison Plot**

Bar Plot:summarizes Profit by state

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='State', y='Profit', data=dc,estimator=sum)
plt.title('Total Profit by State')
plt.show()
```

# **Univariate Composition Plot**

# **5.Perform any probability calculation**

```python
import pandas as pd

data['R&D Spend']=pd.to_numeric(data['R&D Spend'],errors='coerce')
data['Administration']=pd.to_numeric(data['Administration'],errors='coerce')
data['Marketing Spend']=pd.to_numeric(data['Marketing Spend'],errors='coerce')
data['Profit']=pd.to_numeric(data['Profit'],errors='coerce')

total=len(data)
prds= len(data[data['R&D Spend']>150000])/total
pa=len(data[data['Administration']>100000])/total
ps=len(data[data['Marketing Spend']>200000])/total
pp=len(data[data['Profit'] > 200000])/total

print(f"Probability of R&D Spend greater than 150000:{prds:.2f}")
print(f"Probability of Administration greater than 100000:{pa:.2f}")
print(f"Probability of Marketing Spend greater than 200000:{ps:.2f}")
print(f"Probability of Profit greater than 200000:{pp:.2f}")
```
