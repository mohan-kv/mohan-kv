```python
import pandas as pd

data=pd.read_csv('1000_Companies.csv')
print(data)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data[['R&D Spend','Administration','Marketing Spend','Profit']] = data[['R&D Spend','Administration','Marketing Spend','Profit']].apply(pd.to_numeric,errors='coerce')
dc= data.dropna(subset=['R&D Spend', 'Administration', 'Marketing Spend', 'Profit'])

covm= dc[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].cov()
plt.figure(figsize=(10, 8))
sns.heatmap(covm, annot=True,cmap='coolwarm',center=0,linewidths=0.5)
plt.title('Heatmap of Covariance Matrix')
plt.show()
```
