```python
from scipy import stats
import numpy as np

group1 = np.array([23, 25, 21, 28, 22])
group2 = np.array([20, 22, 24, 26, 21])

mean1 = np.mean(group1)
mean2 = np.mean(group2)
std1 = np.std(group1)
std2 = np.std(group2)

print("Mean of Group 1:", mean1)
print("Mean of Group 2:", mean2)
print("Standard Deviation of Group 1:", std1)
print("Standard Deviation of Group 2:", std2)

t_stat, p_value = stats.ttest_ind(group1, group2)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

print("Reject null hypothesis - means are significantly different" if p_value < alpha else "Fail to reject null hypothesis - means are not significantly different")
```
